### Data analysis

```python
#attrition column percentage pie chart
attrition_labels = ['Yes', 'No']
attrition_data = [237, 1233]
# The proportionate percentage is displayed inside the respective wedge with the help of autopct parameter which is set to %1.2f%.
plt.pie(attrition_data, labels = attrition_labels,autopct='%1.2f%%')

plt.title('percentage employee attrition')
plt.show()
```

```python
#income analysis
fig,ax = plt.subplots(3,2, figsize=(20,20))               
plt.suptitle("income factors of attrition", fontsize=20)
sns.boxplot(x=attrition['Attrition'], y=attrition['DailyRate'], ax = ax[0,0])
sns.boxplot(x=attrition['Attrition'], y=attrition['MonthlyIncome'], ax = ax[0,1]) 
sns.boxplot(x=attrition['Attrition'], y=attrition['DistanceFromHome'], ax = ax[1,0]) 
sns.boxplot(x=attrition['Attrition'], y=attrition['MonthlyRate'], ax = ax[1,1]) 
sns.boxplot(x=attrition['Attrition'], y=attrition['DailyRate'], ax = ax[2,0]) 
sns.boxplot(x=attrition['Attrition'], y=attrition['TotalWorkingYears'], ax = ax[2,1])
plt.show()
#Distance from home seems to be a significant factor in employee attrition
```

### 1. Show a breakdown of distance from home by job role and attrition

```python
chart = sns.boxplot(hue=attrition['Attrition'], y=attrition['DistanceFromHome'], x=attrition['JobRole']) 
chart.set_xticklabels(chart.get_xticklabels(), rotation=45)
#The research director and the manufacturing managers had the highest attrition rate when the disance from home was far
```

### 2. Compare average monthly income by education and attrition

```python
chart = sns.boxplot(hue=attrition['Attrition'], y=attrition['MonthlyIncome'], x=attrition['Education'])
chart.set_xticklabels(chart.get_xticklabels(), rotation=45)
```
